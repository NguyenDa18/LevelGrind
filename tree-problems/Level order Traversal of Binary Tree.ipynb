```python
from binarytree import Node
```

```python
root = Node(100)
root.left = Node(50)
root.right = Node(200)
root.left.left = Node(25)
root.left.right = Node(75)
root.right.right = Node(350)

root
```

```python
print(root.right.left == None)
```

```python
class Solution:
    def level_order_traversal(self, root):
        if root == None:
            return None
        print(root.val)
        while root.left != None and root.right != None:
            level_order_traversal(root.left)
            level_order_traversal(root.right)
        return None
```

```python
solution = Solution()
solution.level_order_traversal(root)
```
